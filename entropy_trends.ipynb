# Entropy Trends
In this `JupyterLab` notebook, we explore the general trends related to `Entropy`.

## Setup
First we need to *import* some necessary modules.

```python
# get custom libs
from packages import entropy
from packages import visualize
```

## Visualizing: Entropy vs. # of States
Here we are going to generate some data related to `Entropy` as a function of the number of `states` in whatever system we are interested in. For example a *coin* would involve *two states*, and a *6 sided die* would involve *six states* ... Here we will be looking at systems with states from: $$1 ... 100$$

```python
# first generate entropy dictionary
entropy_states = dict((r, entropy.uniform_entropy(r)) for r in range(1, 100))
```

Now we will use a simple *line plot* to show the *general trend* that *Entropy* follows as the *number of states* increases:
$$
Entropy \propto States
$$

```python
# now plot states and entropy
visualize.line_graph(list(entropy_states.keys()), 
                     list(entropy_states.values()), 
                     '# States:Entropy:', (10, 10));
```

## Visualizing: Entropy vs. Event Probabilities
Now instead of looking at `# of states` we will look at `events` with different *probabilities* (all summing to **1** naturally).

```python
# first we get probability distribution / entropy pairs 
ent_prob = dict((':'.join(map(str, pdst)), entropy.calc_entropy(pdst)) for pdst in entropy.gen_pdist(2, 2))
```

Now we will plot *entropy* vs. different *probability distributions* and see the trend.

```python
# now plot states and entropy
visualize.line_graph(list(ent_prob.keys()), 
                     list(ent_prob.values()), 
                     'Probability Distribution:Entropy:', (10, 10), 
                     tick_labels=True);
```
