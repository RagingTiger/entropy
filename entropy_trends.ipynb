## Visualizing: Entropy vs. # of States
Here we are going to generate some data related to `Entropy` as a function of the number of `states` in whatever system we are interested in. For example a *coin* would involve *two states*, and a *6 sided die* would involve *six states* ... Here we will be looking at systems with states from: $$1 ... 100$$

```python
# first generate entropy dictionary
entropy_trend = dict((r, entropy.gen_entropy(r)) for r in range(1, 100))
```

Now we will use a simple *line plot* to show the *general trend* that *Entropy* follows as the *number of states* increases. There is some noise in the data, the populations for each state in a set of states is *randomly* generated using `random.randint` (hence the noise in the data). Still, the *trend* is clear:
$$
Entropy \propto States
$$

```python
# now plot states and entropy
visualize.line_graph(list(entropy_trend.keys()), 
                     list(entropy_trend.values()), 
                     '# States:Entropy:', (10, 10));
```
